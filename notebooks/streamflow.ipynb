```python
import numpy as np
import os
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = (12, 6)
```

```python
data = np.genfromtxt(os.path.expanduser("Stream Cross Section - Cross section.csv"), delimiter=',')
data = data[~np.any(np.isnan(data), axis=1)]
```

```python
dists = data[:,0]
depths = data[:,1]
```

```python
rel_depth = 17
real_depth = np.min(depths) + rel_depth
assert real_depth < np.max(depths)
```

```python
plt.plot(dists, depths, '-')

plt.axhline(y=0, color='g')
plt.axhline(y=real_depth, color='c')

plt.title("Stream Cross Section")
plt.xlabel("Distance Across [ft]")
plt.ylabel("Depth [cm]")
plt.grid()
```

```python
pos_depths = -depths + real_depth
where_pos = pos_depths >= 0
vol_dists = dists[where_pos]
vol_depths = pos_depths[where_pos]
```

```python
plt.plot(dists, pos_depths)
plt.axhline(y=0, color='g')
plt.grid()
```

```python
plt.plot(vol_dists, vol_depths)
plt.axhline(y=0, color='g')
plt.grid()
```
